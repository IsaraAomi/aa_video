```python
import numpy as np
import glob
import re
import os
import cv2
%matplotlib inline
import matplotlib.pyplot as plt

from opencv_video_to_still import *


def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]

def get_image_flist(dir):
    """
    - Args:
        - dir (str)
    - Returns:
        - flist (list)
    """
    flist = sorted(glob.glob(os.path.join(dir, '*.png')), key=natural_keys)
    return flist

def main():

    image_flist = get_image_flist('../data/image/Shortcake_SONG_shorts_1080pFHR')
    # print(image_flist)

    image_path=image_flist[500]
    print(image_path)
        
    img_color_cv2 = cv2.imread(image_path)
    img_gray_cv2 = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img_color_plt = cv2.cvtColor(img_color_cv2, cv2.COLOR_BGR2RGB)
    plt.imshow(img_color_plt)
    # plt.imshow(img_gray_cv2, cmap='gray')


if __name__ == '__main__':
    main()
```
